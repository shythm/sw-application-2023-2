- S.M.A.R.T. 데이터에서 failure 발생 여부는 다음과 같이 정리할 수 있다.
  - Threshold Exceeded: 특정 값이 임계치를 넘어섰을 때
  - Threshold Not Exceeded: 특정 값이 아직 임계치를 넘어서지 않았을 때
  - Threshold는 S.M.A.R.T. 데이터의 특정 값에 대해 정해진 값이다.
- 일부 Sector를 못 읽는다 하더라도, 이는 failure로 분류되지 않는다.
- S.M.A.R.T. Attributes들은 ATA 표준에 따라 정의되어 있다. 그러나 이 표준은 제조사마다 다르게 구현될 수 있다.
- raw 값 말고, normalized된 값을 보자. normalized 값은 1바이트로, 보통 100이 기본값이며, 값이 높을 수록 좋다. (raw 값은 제조사마다 다르다.)

일단은, 우리가 개발하려는 애플리케이션은 HDD 추천 시스템이다.
HDD를 추천하기에 앞서 HDD의 실패 원인을 파악해보자.
실패 원인을 파악하여, failure 나지 않은 HDD까지도 failure 확률을 계산해볼 수 있을 것이다. 그러면 failure 확률이 높은 HDD에 대해서는
사용 시간에 따라 분석도 필요할 것이다. 일반적인 가정에서와 기업 워크스테이션에서의 사용 시간은 절대적인 영향을 받을 것이기에. (만약, 사용 시간에 대해 아무런 상관 관계가 없다면 굳이?)
failure이 발생한 HDD의 특성 T에 대해 양 또는 음의 상관관계를 가지면 해당 특성 T가 failure에 영향을 미친다는 것을 알 수 있다.
그렇다면, failure이 발생하지 않은 HDD의 특성 T에 대해 상관관계를 가지고 있다면,,, 보다는 failure이 발생하지 않은 HDD의 특성에 대해서 알 필요가 있을까?
failure의 정의가 뭐지? - 특정 값이 임계치를 넘었을 때를 말한다. 이는 제조사가 정한 기준 아닌가?

- S.M.A.R.T. 데이터로부터 failure을 예측한다는 것
  - 제조사마다 그 기준이 "이미" 정해져 있다. 운영체제에서 S.M.A.R.T. 분석을 해보면 최악값과 현재값 그리고 임계값이 존재하는데, 이 임계값이 공장 출하할 때 사전 설정되는 것. (서로 다른 두 하드디스크의 S.M.A.R.T. 정보 비교 사진)
  - S.M.A.R.T. 데이터로부터는 사용 시간, 온도와 같은 사용 패턴에 따른 분석에 이용해야 할 것. 이로부터 failure을 예측한다는 것은 이미 정답을 알고 있는 문제를 또 푸는 것(왜냐하면 임계치를 넘는 특성 T가 존재한다면 이는 failure이기 때문, 그리고 우리는 그 T와 그 임계치를 이미 알고 있음).

failure은 어떤 특성 T가 임계치를 넘었을 때 발생하는 상태이다. 즉, S.M.A.R.T. 기능에는 자체적으로 failure 예측 기능이 들어가 있다는 말이다.
0x01 - Read Error Rate(Vendor Specific Raw Value)
0x05 - Reallocated Sectors Count(드라이브의 수명과도 직결돼있는 특성, 이 또한 제조사에서 정의한 특정 임계치가 있음)
0x0A - Spin Retry Count
